```python
# List of columns to convert
columns_to_convert = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']

# Remove commas and convert to numeric
Vehicle_Data[columns_to_convert] = Vehicle_Data[columns_to_convert].replace({',': ''}, regex=True)
```

```python
Vehicle_Data.tail()
```

```python
#Changing the Data Types
Vehicle_Data[columns_to_convert] = Vehicle_Data[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
Vehicle_Data.isna().sum()
```

```python
# Use pd.melt to reshape the DataFrame
reshaped_df = pd.melt(Vehicle_Data, id_vars=['Vehicle Class', 'Year'], var_name='Month', value_name='Sale')
```

```python
# Sort the reshaped DataFrame by 'Vehicle Class', 'Year', and 'Month'
reshaped_df = reshaped_df.sort_values(by=['Vehicle Class', 'Year', 'Month'])
```

```python
# Reset index for clean output
reshaped_df = reshaped_df.reset_index(drop=True)
```

```python
reshaped_df
```

```python
data={}
for j in range(6):
    temp=[]
    df=pd.DataFrame(columns=['Date', 'Sales'])
    for i in Vehicle_Data[Vehicle_Data["Vehicle Class"]==j].values:
        temp={}
# pd.to_datetime(Vehicle_Data_Final['Year'].astype(str) + '-' + Vehicle_Data_Final['Month'].astype(str) + '-01')
        temp["Date"]=pd.to_datetime([f'{i[1]}-{k}-01' for k in ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN',
       'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']])
        temp["Sales"]=i[2:]
        temp=pd.DataFrame(temp)
        df=pd.concat([df,temp],axis=0,ignore_index=True)
    data[f"{j}"]=df
```

```python
for i in range(6):
    data[f"{i}"]=data[f"{i}"].iloc[:-3,:]
```

# Class 0 [ADAPTED VEHICLE]

```python
data["0"]
```

```python
# Define the number of observations (nobs)
nobs = 12

# Creating training set for 'Date' (independent variable) for class '0'
df_train_X_0 = np.array(data["0"]["Date"].iloc[:-nobs]).reshape(-1,1)

# Creating training set for 'Sales' (dependent variable) for class '0' and converting to integer
df_train_Y_0 = np.array(data["0"]["Sales"].iloc[:-nobs]).reshape(-1,1).astype('int')

# Creating test set for 'Date' (independent variable) for class '0'
df_test_X_0 = np.array(data["0"]["Date"].iloc[-nobs:]).reshape(-1,1)

# Creating test set for 'Sales' (dependent variable) for class '0' and converting to integer
df_test_Y_0 = np.array(data["0"]["Sales"].iloc[-nobs:]).reshape(-1,1).astype('int')

# Checking the shape of the training set for 'Date' in class '0'
df_train_X_0.shape
```

```python
# Importing TimeseriesGenerator from Keras for creating time series data
from keras.preprocessing.sequence import TimeseriesGenerator

# Defining the number of inputs (timesteps) and features for the generator
n_inputs = 12
n_features = 1

# Creating the TimeseriesGenerator for the training data
# This generator will create sequences of 'n_inputs' length from the 'df_train_Y_0' data
generator = TimeseriesGenerator(df_train_Y_0, df_train_Y_0, length=n_inputs, batch_size=1)

# Iterating through the generator to extract and print the input sequences (X) and corresponding output (y)
for i in range(len(generator)):
    X, y = generator[i]
    print(f'\nInput Sequence: {X.flatten()} and Target Value: {y}')
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import GRU

model = Sequential()
model.add(GRU(300, activation='relu', input_shape=(n_inputs, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.summary()
```

```python
# Training the model using the TimeseriesGenerator data
model.fit(generator, epochs=300, verbose=0)
```

```python
import matplotlib.pyplot as plt
plt.plot(model.history.history['loss'])
```

```python
model.predict([df_test_Y_0])
```

```python
df_test_Y_0
```

```python
# plotting actual vs predicted
plt.scatter(df_test_X_0,df_test_Y_0,label="Original Label")
plt.scatter(df_test_X_0,model.predict([df_test_Y_0]),label="Prediction Label")
plt.title("Class 0")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
# calculating the errors
print(f"R2 score : {r2_score(df_test_Y_0,model.predict([df_test_Y_0]))}")
print(f"MSE : {mean_squared_error(df_test_Y_0,model.predict([df_test_Y_0]))}")
print(f"RMSE : {(mean_squared_error(df_test_Y_0,model.predict([df_test_Y_0])))**(1/2)}")
```

```python
model.predict(df_test_Y_0.reshape(1,-1))[0][0]
```

```python
# Initialize empty list for storing predictions
temp = []

# Reshape the test data for prediction
Data = df_test_Y_0.reshape(1, -1)[0]

# Generate predictions iteratively
for i in range(3):
    pred = model.predict(np.array([Data]))[0][0]
    temp.append(pred)
    Data = Data[1:]
    Data = np.append(Data, pred)
```
